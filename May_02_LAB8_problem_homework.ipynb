### LAB 8-1 오늘의 날짜와 현재 시간 출력하기

datetime 모듈을 사용하여 현재 날짜와 시간을 가져오고 년, 월, 일을 따로 출력하고, 시간은 오전/오후를 구분해서 출력하여라.

```python
import datetime
now = datetime.datetime.now()

year = now.year
month = now.month
day = now.day
hour = now.hour
minute = now.minute
second = now.second
if hour < 12:
    print(f"오늘의 날짜 : {year}년 {month}월 {day}일")
    print(f"현재시간 : 오전 {hour}시 {minute}분 {second}초")
else:
    print(f"오늘의 날짜 : {year}년 {month}월 {day}일")
    print(f"현재시간 : 오후 {hour-12}시 {minute}분 {second}초")
```

### LAB 8-2 남은 날짜 계산하기


datetime 모듈을 이용하여현재 날짜를 기준으로 특정 날짜까지 남은 날짜와 시간을 계산하여라.


```python
import datetime
now = datetime.datetime.now()

year = now.year
month = now.month
day = now.day
hour = now.hour
print(f"오늘은 {year}년 {month}월 {day}일입니다.")

Christmas = datetime.datetime(2027, 12, 25)
diff1 = Christmas - now

print(f"2027년 '크리스마스'까지는 {diff1.days}일 {diff1.seconds//3600}시간 남았습니다.")

New_year = datetime.datetime(2036, 1, 1)
diff2 = New_year - now

print(f"2036년 '새해'까지는 {diff2.days}일 {diff2.seconds//3600}시간 남았습니다.")

birthday = datetime.datetime(2026, 10, 13)
diff3 = birthday - now

print(f"'생일'까지는 {diff3.days}일 {diff3.seconds//3600}시간 남았습니다.")
```

### LAB 8-3 날짜 계산하기

datetime을 이용하여 오늘 날짜에서 특정 일수 이후의 날짜를 구하고, 입력한 날짜 기준으로 기념일을 계산하여라.

```python
import datetime
now = datetime.datetime.now()
after_1000 = now + datetime.timedelta(days=1000)
print(f"오늘 날짜 : {now.year}년 {now.month}월 {now.day}일")
print(f"1000일 후 : {after_1000.year}년 {after_1000.month}월 {after_1000.day}일")
```

```python
y, m, d = map(int, input("처음 사귄 날짜 입력 (년 월 일): ").split())
start_day = datetime.datetime(y, m, d)
day100 = start_day + datetime.timedelta(days=100)
print(f"100일 기념일은 : {day100.year}년 {day100.month}월 {day100.day}일입니다.")
```

### LAB 8-5 랜덤 번호 생성기

random 모듈을 사용하여 임의의 숫자를 생성하고 randrange()와 sample()을 이용하여 조건에 맞는 숫자들을 선택하여라.

```python
import random
result1 = []
while len(result1) < 3:
    num = random.randrange(0, 101)
    if num % 5 == 0:
        result1.append(num)
print("0에서 100 이하의 정수 중에서 5의 배수:", result1)
```

```python
import random
result2 = random.sample(range(1, 11), 3)
print("1에서 10 사이의 임의의 정수:", result2)
```

### LAB 8-6 로또 번호 얻기

random 모듈을 사용하여 로또 번호를 생성하고 중복 없는 번호를 생성하여 여러 개의 로또 번호와 보너스 번호를 출력한다.

```python
import random as rd
print("이번 주의 추천 로또 번호")
for _ in range(5):
    lotto_list = list(range(1, 46))
    rd.shuffle(lotto_list)
    lotto = lotto_list[:6]
    lotto.sort()
    print(lotto)
```

```python
import random as rd
lotto_list = list(range(1, 46))
rd.shuffle(lotto_list)
main = lotto_list[:6]
bonus = lotto_list[6]
main.sort()
print("이번 주의 당첨 로또 번호 + 보너스")
print(f"{main} + [{bonus}]")
```

### 연습문제 8.3 함수 수행 시간 비교

여러 가지 함수를 정의하고 각각을 100번 반복 실행하여 수행 시간을 측정하고 각 함수의 실행 시간을 비교하여라.

```python
import time
def hello():
    print("Hello Python!")
start = time.time()
for _ in range(100):
    hello()
end = time.time()
print(f"Hello Python! 출력을 100번 반복하는 시간 : {end - start:.4f}초")
```

```python
import time
def factorial():
    result = 1
    for i in range(1, 10001):
        result *= i
    return result
start = time.time()
for _ in range(100):
    factorial()
end = time.time()
print(f"10000!을 100번 반복해서 구하는 시간 : {end - start:.4f}초")
```

```python
import time
def sum_odd_cubes():
    total = 0
    for i in range(1, 10001, 2):
        total += i**3
    return total
start = time.time()
for _ in range(100):
    sum_odd_cubes()
end = time.time()
print(f"1에서 360도까지 sin 값의 합을 100번 반복하는 시간 : {end - start:.4f}초")
```

```python
import time
import math
def sum_sin():
    total = 0
    for i in range(1, 361):
        total += math.sin(math.radians(i))
    return total

start = time.time()
for _ in range(100):
    sum_sin()
end = time.time()

print(f"1에서 360도까지 sin 값의 합을 100번 반복하는 시간 : {end - start:.4f}초")
```

### 연습문제 8.4 의사난수 생성기

의사난수를 생성하는 함수 my_rand()를 구현하여라. time 모듈을 이용하여 초기 시드를 설정하고, 0부터 1,000,000 사이의 난수를 생성하여라.

```python
seed = 1
def my_rand():
    global seed
    seed = (1103515245 * seed + 12345) % (2**31)
    return seed % 1000001
print("의사난수 10개")
for _ in range(10):
    print(my_rand())
```
