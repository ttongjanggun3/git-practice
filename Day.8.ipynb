```python
dog = {}
dog['name'] = 'Choco'
dog['color'] = 'Brown'
dog['breed'] = 'Poodle'
dog['legs'] = 4
dog['age'] = 3
print("dog:", dog)

student = {
    'first_name': 'GyuWon',
    'last_name': 'Kim',
    'gender': 'Male',
    'age': 22,
    'marital_status': 'Single',
    'skills': ['Python', 'C'],
    'country': 'Korea',
    'city': 'Daegu',
    'address': 'Somewhere'
}

print("student:", student)
print("student 길이:", len(student))

print("skills:", student['skills'])
print("type:", type(student['skills']))
student['skills'].append('Java')
student['skills'].append('Git')
print("skills 수정:", student['skills'])
print("keys:", list(student.keys()))
print("values:", list(student.values()))
print("items:", list(student.items()))
student.pop('address')
print("address 삭제 후:", student)
del dog
print("dog 삭제 완료")
```
